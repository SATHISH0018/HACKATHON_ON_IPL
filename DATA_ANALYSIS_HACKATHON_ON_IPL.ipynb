# 1. Which of the following is the right way to merge 2 datasets for getting better insights from the data

```python
matches.rename(columns={'id':'match_id'},inplace=True)
```

### Data Cleaning 

```python
data.isna().sum()
```

```python
# Filling the  missing values in the 'city' column with 'NO_CITY'.

data['city'] = data['city'].fillna('NO_CITY')

# Filling the missing values in the 'player_of_match' to 'No_player_of_the_match'.

data['player_of_match'] = data['player_of_match'].fillna('No_player_of_the_match')

# Filling the missing values.

data['winner'] = data['winner'].fillna('no_winner')

# Fill missing values in the 'player_dismissed' column with 'no_player_dismissed'.

data['player_dismissed'] = data['player_dismissed'].fillna('no_player_dismissed')

# Fill missing values in the 'dismissal_kind' column with 'no_dismissal_kind'.

data['dismissal_kind'] = data['dismissal_kind'].fillna('no_dismissal_kind')

# Fill missing values in the 'fielder' column with 'no_fielder'

data['fielder'] = data['fielder'].fillna('no_fielder')
```

```python
data.isna().sum()
```

```python
# dropping umpires columns based on a problem statement.

umpire_columns_to_drop = ['umpire2', 'umpire3']
data.drop(columns=umpire_columns_to_drop,inplace=True)
```

```python
data.info()
```

```python
# changing date column 

data['year'] = pd.to_datetime(data['date'])
data['year']=(data['year']).dt.year
```

# 12. Most number of wickets taken by a bowler is


```python
wkt = deliveries.loc[(deliveries['batsman'] == deliveries['player_dismissed']) & (deliveries['dismissal_kind']!= 'run out')]
wkt
```

```python


top_bowlers = wkts.sort_values('inning',ascending=False).head(5)

top_bowlers.plot(kind='bar', color='blue')
plt.title('Top Bowlers by Total Number of Wickets')
plt.xlabel('Bowler')
plt.ylabel('Total Wickets')
plt.show()

```
